##### Date Feature Extraction
Untuk mempermudah perwakilan informasi pada tanggal, akan ditambahkan fitur tahun, bulan, hari dalam minggu, dan hari dalam bulan pada setiap fitur bertipe tanggal (Order date dan Ship Date).

```python
class DateHandler(BaseEstimator, TransformerMixin):
   def __init__(self, isNeeded = True): # no *args or **kargs
      self.isNeeded = isNeeded
   def fit(self, X, y=None):
      return self # nothing else to do
   def transform(self, X, y=None):
      X_handled = X.copy()
      if (self.isNeeded):
         X_handled['Order_year'] = X_handled['Order Date'].dt.year
         X_handled['Order_month'] = X_handled['Order Date'].dt.month
         X_handled['Order_dayInMonth'] = X_handled['Order Date'].dt.days_in_month
         X_handled['Order_dayofWeek'] = X_handled['Order Date'].dt.day_of_week

         X_handled['Month_Between'] = (X_handled['Ship Date'].dt.year - X_handled['Order Date'].dt.year) * 12 + (X_handled['Ship Date'].dt.month - X_handled['Order Date'].dt.month)
      X_handled.drop('Ship Date', axis=1, inplace=True)
      X_handled.drop('Order Date', axis=1, inplace=True)
      return X_handled
```

##### Clustering with K-Means and Checking Outlier with DBSCAN

Pada tahap ini, dilakukan clustering customer menggunakan K-Means untuk menentukan customer persona. Penentuan jumlah cluster akan ditentukan berdasarkan grafik Cumulative Variance vs Number of Components dengan dimensionality reduction menggunakan PCA. Sebelum melakukan clustering, saya akan mengecek outlier dengan DBSCAN.

Saya menghapus beberapa fitur yang mungkin tidak perlu untuk clustering.

```python
D_Cluster = X_train.copy()
D_Cluster.drop(['Product Name', 'Order Date', 'Ship Date'], axis=1, inplace=True)
```

```python
### Label encoding ###
LABELS = D_Cluster.columns

encoder = LabelEncoder()

for col in LABELS:
    # Check if object
    if D_Cluster[col].dtype == 'O':
        # Fit label encoder and return encoded labels
        D_Cluster[col] = encoder.fit_transform(D_Cluster[col]) 
```

```python
### Scaling ###
mm_scaler = MinMaxScaler() 

# Apply Min-Max Scaling
D_Cluster[LABELS] = mm_scaler.fit_transform(D_Cluster[LABELS])

D_Cluster.head()
```

```python
# Fit your PCA
pca_custom = PcA(n_components=10)
pca_custom.fit(D_Cluster)

# The amount of variance that each PC explains (custom implementation)
var_custom = np.cumsum(np.round(pca_custom.explained_variance_ratio_, decimals=4) * 100)
print("Variance from custom model: ", var_custom)

# Fit sklearn PCA
pca_sklearn = PCA(n_components=10)
pca_sklearn.fit(D_Cluster)

# The amount of variance that each PC explains (scikit-learn)
var_sklearn = np.cumsum(np.round(pca_sklearn.explained_variance_ratio_, decimals=4) * 100)
print("Variance from sklearn model: ", var_sklearn)

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot for custom PCA
ax1.plot(var_custom, linewidth=3)
ax1.set_xlabel('Number of components', fontsize=15)
ax1.set_ylabel('Cumulative explained variance', fontsize=15)
ax1.set_title('Custom PCA', fontsize=20)
ax1.set_xticks(np.arange(1, len(var_custom) + 1, step=1))
ax1.set_facecolor('lightgoldenrodyellow')

# Plot for scikit-learn PCA
ax2.plot(var_sklearn, linewidth=3)
ax2.set_xlabel('Number of components', fontsize=15)
ax2.set_ylabel('Cumulative explained variance', fontsize=15)
ax2.set_title('Scikit-learn PCA', fontsize=20)
ax2.set_xticks(np.arange(1, len(var_sklearn) + 1, step=1))
ax2.set_facecolor('lightgoldenrodyellow')

# Overall figure settings
fig.suptitle('Comparison of Custom PCA and Scikit-learn PCA', fontsize=25)
plt.tight_layout()
plt.show()
```

Dari grafik dan nilai variansi di atas, saya memilih N = 9 karena masih mewakili lebih dari 95% dari total variansi

```python
# Initiating PCA to reduce dimentions to 14
pca = PcA(n_components=9)
pca.fit(D_Cluster) 
D_pca = pca.transform(D_Cluster)
```

Outlier Detection

Dari grafik dapat dilihat nilai point naik secara tajam di distance sekitar 2.5. Oleh karena itu, saya menggunakan nilai epsilon = 2.5.

Clustering

Untuk menentukan jumlah cluster, saya menggunakan elbow method.

```python
inertia_custom = []
for i in range(1,9):
    kmeans = Kmeans(n_clusters=i,init='k-means++', random_state=42)
    kmeans.fit(D_pca)
    inertia_custom.append(kmeans.inertia_)
```

```python
inertia_sklearn = []
for i in range(1, 9):
    kmeans_sklearn = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans_sklearn.fit(D_pca)
    inertia_sklearn.append(kmeans_sklearn.inertia_)
```

```python
# Plotting the inertia for both implementations
fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(range(1, 9), inertia_custom, marker='o', linestyle='-', label='Custom KMeans')
ax.plot(range(1, 9), inertia_sklearn, marker='x', linestyle='--', label='Scikit-learn KMeans')

ax.xaxis.set_tick_params(labelsize=13)
ax.yaxis.set_tick_params(labelsize=13)
ax.set_xlabel("Number of Clusters", fontsize=15)
ax.set_ylabel('Inertia', fontsize=15)
ax.set_title("Elbow Method Comparison", fontsize=22)
ax.legend(fontsize=13)

plt.tight_layout()
plt.show()
```

Berdasarkan grafik elbow method, saya belum bisa menyimpulkan dengan yakin berapa jumlah cluster (N) yang dibutuhkan. </p>
Oleh karena itu, saya juga menghitung silhouette score.

```python
range_clusters = range(2,9)
silhouette_scores = []
for i in range_clusters:
    kmeans = Kmeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(D_pca) 
    labels = kmeans.predict(D_pca) 
    score = silhouette_score(D_pca,labels)
    silhouette_scores.append(score)
```

```python
plt.plot(range_clusters,silhouette_scores,marker='o')
plt.xlabel("number of clusters",fontsize=15)
plt.ylabel('silhouette score',fontsize=15)
plt.title("Silhouette Score",fontsize=22)

plt.tight_layout()
plt.gca()
plt.show()
```

Nilai tertinggi silhouette score terdapat di jumlah cluster 3. Oleh karena itu, saya memutuskan untuk membagi customer menjadi 3 cluster/persona.

```python
k_means = Kmeans(n_clusters=3, init="k-means++", random_state=32)
cluster_feature = k_means.fit_predict(D_pca)
D_pca['Cluster'] = cluster_feature
X_train['Cluster'] = cluster_feature
X_EDA['Cluster'] = cluster_feature
```

### Data Preparation

```python
class RemoveOutliers(BaseEstimator, TransformerMixin):
   def __init__(self, isNeeded=True):
      self.isNeeded = isNeeded
   def fit(self, X, y=None):
      return self # nothing else to do
   def transform(self, X, y):
      X_removed = X.copy()
      if self.isNeeded:
         columns = ['Sales', 'Quantity', 'Discount']
         Q1 = X[columns].quantile(0.10)
         Q3 = X[columns].quantile(0.90)
         IQR = Q3 - Q1
         
         # Define the upper and lower bounds for outlier detection
         lower_bound = Q1 - 1.5 * IQR
         upper_bound = Q3 + 1.5 * IQR
         
         # Remove outliers from each specified column
         for col in columns:
            X_removed = X_removed[(X_removed[col] >= lower_bound[col]) & (X_removed[col] <= upper_bound[col])]
      
      return X_removed
```

```python
def label_encode_by_mean_profit(X, cat_attribs, m=10):
    overall_mean = X['Profit'].mean()  # Compute overall mean profit
    encoding_list = []  # Dictionary to store encoding maps for each categorical attribute

    for col in cat_attribs:
        # Compute mean profit and count for each category
        category_stats = X.groupby(col)['Profit'].agg(['mean', 'count'])
        
        # Apply smoothing formula
        smoothed_means = (category_stats['mean'] * category_stats['count'] + overall_mean * m) / (category_stats['count'] + m)
        
        # Create a mapping from category to smoothed mean profit
        encoding = smoothed_means.to_dict()
        
        # Store the mapping in the encoding dictionary
        encoding_list.append(encoding)
    
    return encoding_list

encoding_list = label_encode_by_mean_profit(train, cat_attribs)
```

```python
class TargetEncoding(BaseEstimator, TransformerMixin):
   def __init__(self, encoding_list, cat_attribs, overall_mean):
      self.encoding_list = encoding_list
      self.cat_attribs = cat_attribs
      self.overall_mean = overall_mean
   def fit(self, X, y=None):
      return self # nothing else to do
   def transform(self, X, y=None):
      for idx, col in enumerate(self.cat_attribs):
        X[col] = X[col].map(encoding_list[idx])
        if (X[col].isna().sum() > 0):
           X[col].fillna(self.overall_mean, inplace=True)
      return X
```

Karena fitur kategori City, State, dan Product Name memiliki jenis yang banyak (> 1000) maka saya menggunakan smoothen target encoding berdasarkan rata-rata profit/laba untuk setiap kategori. Smoothen yang dimaksud adalah menghaluskan nilai profit pada suatu kategori berdasarkan perbandingan jumlah kemunculanya dengan nilai m. Semakin sedikit kemunculan suatu kategori, semakin besar pengaruh nilai rata-rata profit keseluruhan terhadap nilai pengganti kategori tersebut. Hal ini untuk mencegah overfitting karena sedikitnya keberadaan suatu kategori.

```python
class ClassHandler(BaseEstimator, TransformerMixin):
   def __init__(self, encoding_list, cat_attribs, overall_mean, isTargetEncoding = False):
      self.isTargetEncoding = isTargetEncoding
      self.encoding_list = encoding_list
      self.cat_attribs = cat_attribs
      self.overall_mean = overall_mean
   def fit(self, X, y=None):
      if not self.isTargetEncoding:
         self.one_hot_encoder = OneHotEncoder(handle_unknown='ignore')
         self.one_hot_encoder.fit(X)
      return self
   def transform(self, X, y=None):
      if (self.isTargetEncoding):
         X_encoded = TargetEncoding(self.encoding_list, self.cat_attribs, self.overall_mean).fit_transform(X)
            
         # Scale the features
         scaler = StandardScaler()
         X_scaled = scaler.fit_transform(X_encoded)
         
         return X_scaled
      else:
         return self.one_hot_encoder.transform(X)
```

Meskipun onehot encoder normalnya akan memakan waktu lebih lama (tapi nanti akan memakai sparse matrix jadi mungkin waktu kalkulasi tidak menjadi masalh), bisa jadi onehot encoder akan mempunyai hasil lebih baik karena tidak prone terhadap overfitting layaknya target encoding. Oleh karena itu, saya mencoba keduanya dengan mengatur hyperparameter isTargetEncoding pada kelas di atas. </p>
Dibutuhkan standard sclaer setelah target encoding agar dapat diolah dengan baik oleh model yang rentan terhadap normalitas data, seperti polynomial regression. (Sebelum dikasih scaler, polynomial pake target encoding kacau :v)

```python
num_pipeline = Pipeline([
 ('std_scaler', StandardScaler()),
 ('power_transfrom', PowerTransformer(method='yeo-johnson'))
 ])

numCat_pipeline = ColumnTransformer([
 ("num", num_pipeline, num_attribs),
 ("cat", ClassHandler(encoding_list, cat_attribs, y_train.mean(), isTargetEncoding=True), cat_attribs),
 ])

full_pipeline = Pipeline([
    ('Handling Date', DateHandler()),
    # ('Remove Outlier', RemoveOutliers(isNeeded=True)),
    ('NumCat Pipeline', numCat_pipeline)
])
```

Ditambahkan standard scaler pada fitur angka untuk mengurangi volatilitas fitur. Power transform digunakan untuk menormalisasi fitur-fitur angka dengan menggunakan metode yeo-johnson karena profit sendiri memiliki distribusi normal.
